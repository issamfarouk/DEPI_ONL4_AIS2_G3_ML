```python
# Task -1:
# Suppose you have this email address “Amit_ml@gmail.edu”
# - Input Validation: Check if the input string contains exactly one "@" symbol and at least one "." after the "@" symbol. If it's not a valid email, return "Invalid email".
# - Extract Username: Extract and return the part of the email before the "@" symbol.
# - Extract Domain: Extract and return the domain (the part between "@" and the last ".").
# - Check for Domain Ending: Check if the email ends with ".com". If it does, return "Commercial Domain". If it ends with ".edu", return "Educational Domain". Otherwise, return "Other Domain".
```

```python
# - Input Validation: Check if the input string contains exactly one "@" symbol and at least one "." after the "@" symbol. If it's not a valid email, return "Invalid email".
email = "Amit_ml@gmail.edu"
if email.count("@") == 1 and "." in email.split("@")[1]:
    print("Valid email")
else:
    print("Invalid email")
```

```python
# - Extract Username: Extract and return the part of the email before the "@" symbol.
user_name = email[0:7]
print(user_name)
```

```python
# - Extract Domain: Extract and return the domain (the part between "@" and the last ".").
start = email.find("@") + 1
end = email.find(".")
domain = email[start:end]
print(domain)
```

```python
# - Check for Domain Ending: Check if the email ends with ".com". If it does, return "Commercial Domain". If it ends with ".edu", return "Educational Domain". Otherwise, return "Other Domain".
if email.endswith(".com"):
    print("Commercial Domain")
elif email.endswith(".edu"):
    print("Educational Domain")
else:
    print("Other Domain")
```

```python
# Task -2:
# Encoded Message: ###!!@mocleW EPGTQ!!!6789
# Steps to Decode:
# 1. Extract the core part of the message: "mocleW EPGTQ".
# 2. Reverse the first word: "mocleW" becomes "Welcome".
# 3. Replace shifted vowels in the second word:
#    o "EPGTQ": No vowels to change.
# 4. Final decoded message: "Welcome PGTQ".
```

```python
# 1. Extract the core part of the message: "mocleW EPGTQ".
message = "###!@mocleW EPGTQ!!!6789"
core_message = message[5:-7]
print(core_message)
```

```python
# 2. Reverse the first word: "mocleW" becomes "Welcome".
first_word = core_message.split(" ")[0]
reversed_first_word = first_word[::-1]
print(reversed_first_word)
```

```python
# 3. Replace shifted vowels in the second word:
#    o "EPGTQ": No vowels to change.
second_word = core_message.split(" ")[1]
print(second_word)
```

```python
# 4. Final decoded message: "Welcome PGTQ".
final_message = reversed_first_word + " " + second_word
print(final_message)
```
